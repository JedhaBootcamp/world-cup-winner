```python
import requests
from bs4 import BeautifulSoup
```

```python
team_wc = requests.get("https://en.wikipedia.org/wiki/All-time_table_of_the_FIFA_World_Cup")
```

```python
soup = BeautifulSoup(team_wc.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
tables = soup.find_all("table")
```

```python
for table in tables:
    print(table)
```
